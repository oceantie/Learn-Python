```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import MultipleLocator,FormatStrFormatter
#刻度
t=np.arange(0.0,100.0,1)
s=np.sin(0.1*np.pi*t)*np.exp(-t*0.01)
ax=plt.subplot(111)#注意，一般都在ax中设置，不再plot中设置
plt.plot(t,s,'--*')
plt.grid(True,linestyle='--',color='grey',linewidth='0.5',axis='both')

xmajorLocator=MultipleLocator(10)#将x主刻度设为10的倍数
xmajorFmatter=FormatStrFormatter('%.0f')
xminLocator=MultipleLocator(5)#将x主刻度设为5的倍数
ymajorLocator=MultipleLocator(0.5)#将x主刻度设为10的倍数
ymajorFmatter=FormatStrFormatter('%.1f')
yminLocator=MultipleLocator(0.1)#将x主刻度设为5的倍数

ax.xaxis.set_major_locator(xmajorLocator)
ax.xaxis.set_major_formatter(xmajorFmatter)
ax.xaxis.set_minor_locator(xminLocator)
ax.yaxis.set_major_locator(ymajorLocator)
ax.yaxis.set_major_formatter(ymajorFmatter)
ax.yaxis.set_minor_locator(yminLocator)

ax.xaxis.grid(True,which='major')#控制主刻度，次刻度网格，只显示主刻度网格
ax.yaxis.grid(True,which='minor')

#删除坐标轴刻度显示
# ax.yaxis.set_major_locator(plt.NullLocator())
# ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#图表输出
df=pd.DataFrame(np.random.randn(1000,4),columns=list('ABCD'))
df=df.cumsum()
df.plot(style='--',alpha=0.5)
plt.legend(loc='upper left')
plt.savefig('F:/Python/pic222.png',
            dpi=400,
            bbox_inches='tight',
            facecolor='g',
            edgecolor='b'
           )
#可支持png,pdf,svg,ps
#bbox_inches:图标需要保存的部分，如果设置为tight，则尝试剪出图表空白部分
#facecolor.edgecolor:图像背景色，默认为‘w’白色
```

```python
#子图
#在matplotlib中，整个图像为一个Figure对象
#在figure对象中可以包含一个或者多个axes对象
#每个axes对象都是一个拥有自己坐标系统的绘图区域

#plt.figure()  绘图对象
#plt.figure(num=None,figsize=None,dpi=None,facecolor=None,edgecoloe=None,
#frameon=True,)

fig1=plt.figure(num=1,figsize=(4,2))
plt.plot(np.random.rand(50).cumsum(),'k--')
fig2=plt.figure(num=2,figsize=(4,2))
plt.plot(50-np.random.rand(50).cumsum(),'k--')

#num:图表序号，表示第几个图表
#figsize：图表大小
#懂我吗用plot时，如果设置plt.figure(),则会调用figure（）生成一个figure，
```

```python
#创建子图1
fig=plt.figure(figsize=(10,6),facecolor='grey')
ax1=fig.add_subplot(2,2,1)#第一行的左图
plt.plot(np.random.rand(50).cumsum(),'k--')
plt.plot(np.random.randn(50).cumsum(),'b--')

ax2=fig.add_subplot(2,2,2)#第一行的右图
ax2.hist(np.random.rand(50),alpha=0.5)

ax4=fig.add_subplot(2,2,4)#第2行的右图
df2=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
ax4.plot(df2,alpha=0.5,linestyle='--',marker='.')
```

```python
#创建子图2
fig,axes=plt.subplots(2,3,figsize=(10,4))
ts=pd.Series(np.random.randn(1000).cumsum())
print(axes,axes.shape,type(axes))
#生成图表对象的数组
ax1=axes[0,1]
ax1.plot(ts)
axes[0,0].plot(np.random.rand(100))
```

```python
#子图创建3
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list('ABCD'))
df=df.cumsum()
df.plot(style='--',alpha=0.4,grid=True,figsize=(8,8),
       subplots=True,
        layout=(2,3),
        sharex=False)
plt.subplots_adjust(wspace=0,hspace=0.2)
#plt.subplots_adjust(wspace=0,hspace=0.2)
#plt.plot()基本图表绘制函数，subplot是否分别绘制子图
#layout:绘制子图矩阵，按序列填充
```

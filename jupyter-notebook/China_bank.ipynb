```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ChinaBank = pd.read_csv('ChinaBank.csv',index_col = 'Date')
ChinaBank = ChinaBank.iloc[:,1:]
ChinaBank.index = pd.to_datetime(ChinaBank.index)
Close = ChinaBank.Close

# print(Close['2014'])
plt.plot(Close['2014'])
plt.show()



# plt.plot([1,1,0,0,-1,0,1,1,-1])
# plt.ylim(-1.5,1.5)
# plt.xticks(range(9),['02-01','02-02','02-03','02-04','02-05','02-06','02-07','02-08','02-09'],rotation=45)
#我们可以通过xticks()和yticks()函数设定坐标的标签，两个函数主要有以下两个参数：
#location:指坐标的位置.
#labels：对应的坐标位置显示的标签.
#两个参数一般是等长的数组,该函数另一个常用的参数是rotation参数，可以对坐标标签进行适当的旋转，看下面的例子：
```

```python
plt.plot(Close['2014'])
plt.title('Bank of China close price in 2014',loc='right')#增加标题
plt.show()
```

```python
plt.plot(Close['2014'])
plt.title('Bank of China close price in 2014',loc='left')#loc为标题位子
plt.xlabel('day')#x轴标签
plt.ylabel('price')
plt.show()
```

```python
plt.plot(Close['2014'])
plt.title('Bank of China close price in 2014',loc='left')#loc为标题位子
plt.xlabel('day')#x轴标签
plt.ylabel('price')
plt.grid(True,axis='y')
plt.show()
```

```python
Open=ChinaBank.Open
plt.plot(Close['2014'],label='close price')
plt.plot(Open['2014'],label='open price')
plt.legend(loc='best')#增加图例使用legend函数，best自动选取最佳位置
plt.show()
```

```python
plt.plot(Close['2014'],label='close price',linestyle='solid')
plt.plot(Open['2014'],label='open price',ls='-.')#ls表示linestyle
plt.legend(loc='best')
plt.xlabel('day')
plt.ylabel('close price')
plt.title('Bank of China close price in 2014',loc='center')
plt.grid(True,axis='y')
plt.show()
```

```python
#颜色设置
plt.plot(Close['2014'],label='close price',c='r',linestyle='solid')
plt.plot(Open['2014'],label='open price',c='g',ls=':')#c表示color
plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('close price')
plt.title('Bank of China close price in 2014',loc='center')
plt.grid(True,axis='y')
plt.show()
```

```python
#图形形状设置
plt.plot(Close['2014'],label='close price',c='r',marker='o',linestyle='solid')
plt.plot(Open['2014'],label='open price',c='b',marker='v',ls='-.')#marker可以设置图形形状
plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('close price')
plt.title('Bank of China close price in 2014',loc='center')
plt.grid(True,axis='y')
plt.show()
```

```python
#线条宽度可以通过plot函数中的linewidth函数指定，也可简写为lw
plt.plot(Close['2014'],label='close price',c='r',marker='o',linestyle='solid',linewidth=9)
plt.plot(Open['2014'],label='open price',c='b',marker='>',ls='-.',lw=3)
plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('close price')
plt.title('Bank of China close price in 2014',loc='center')
plt.grid(True,axis='y')
plt.show()
```

```python
#合并写法
#线条的类型，图形的颜色和点的形状类型，可以合为一个属性,但是必须放在label前面
plt.plot(Close['2014'],'--rD',label='close price',linewidth=1)
plt.plot(Open['2014'],'--b>',label='open price',lw=3)
plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('close price')
plt.title('Bank of China close price in 2014',loc='center')
plt.grid(True,axis='y')
plt.show()
```
